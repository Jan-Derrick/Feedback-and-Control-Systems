# Problem 1

```python
import control as ctrl
import matplotlib.pyplot as plt

num1 = [1]        
den1 = [1, 2, 2]

G = ctrl.tf(num1, den1)
print(G)
```

```python
poles1 = ctrl.poles(G)
print("No. 1 Poles: ", poles1)
```

```python
zeroes1 = ctrl.zeros(G)
print("No. 1 Zeroes: ", zeroes1)
```

```python
ctrl.pzmap(G, grid=True)
plt.show()
```
